```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression, LinearRegression, SGDClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report, precision_recall_curve
from sklearn.neighbors import KNeighborsClassifier, NearestNeighbors
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
```

## Social Network Ads Data

This dataset shows gender, age and approximate annual salary of social network users and whether or not they purchased the product that was advertised.

```python
snads = pd.read_csv('data/Social_Network_Ads.csv')
```

```python
snads.head()
```

```python
snads.info()
```

#### Visualizing how Age and Estimated Salary affects the likelihood that a user will purchase the product advertised using boxplots. 

##### It appears in general, older people purchased products through social media ads. 
##### Additionally, people who earned higher salarys on average purchased more through social media ads.

```python
plt.figure(figsize=(15, 5))
plt.subplot(121)
sns.boxplot(x= 'Purchased', y='Age', data=snads)
plt.subplot(122)
sns.boxplot(x='Purchased', y='EstimatedSalary', data=snads)
```

#### Using a color-coded scatterplot to compare likelihood of purchase comparing Estimated Salary and Age. 

```python
colors = pd.get_dummies(snads.Purchased, drop_first=True)
plt.scatter(snads.Age, snads.EstimatedSalary, c = colors[1], cmap='jet', alpha=0.2)
```

```python
plt.plot(snads.EstimatedSalary, colors[1], 'o')
```

```python
X = snads.EstimatedSalary.values.reshape(-1,1)
y = colors[1]
X_train, X_test, y_train, y_test = train_test_split(X,y)
clf = LogisticRegression()
clf.fit(X,y)
```

```python
clf.intercept_
```

```python
clf.coef_
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(60000)
```

```python
clf.predict_proba(60000)
```

```python
y = clf.predict_proba(snads.EstimatedSalary.values.reshape(-1,1))
```

#### Used a line plot to depict how likely someone who had an Estimated Salary of 65,000 dollars would purchase through a social media advertisement. It appears not likely.

```python
plt.plot(np.sort(snads.EstimatedSalary), np.sort(y[:, 1]))
plt.plot(snads.EstimatedSalary, colors[1], 'o')
plt.plot(65000, clf.predict_proba(65000)[:, 1], 'o', markersize = 12)
clf.predict_proba(65000)[:, 1]
plt.xlabel('Estimated Salary')
plt.ylabel('Probability of Purchase')
```

```python
pipe = make_pipeline(PolynomialFeatures(), LogisticRegression())
```

```python
params = {'polynomialfeatures__degree': [i for i in range (1,6)], 
          'logisticregression__C': [0.1, 1, 5, 10, 20, 50, 100]}
```

```python
grid = GridSearchCV(pipe, param_grid=params, cv=10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(snads.EstimatedSalary.values.reshape(-1,1), colors[1])
```

```python
grid.fit(X_train,y_train)
```

#### According to the grid search, it seems the best coefficient and degree are both 1

```python
grid.best_estimator_
```

```python
grid.best_estimator_.score(X_test,y_test)
```

```python
X = snads.Age.values.reshape(-1,1)
y = snads.Purchased
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
lgr = LogisticRegression()
sgd = SGDClassifier()
```

```python
lgr.fit(X_train, y_train)
```

```python
lgr.score(X_test, y_test)
```

```python
sgd.fit(X_train, y_train)
```

```python
sgd.score(X_test, y_test)
```

```python
dum_dum= DummyClassifier(strategy = 'most_frequent')
```

```python
dum_dum.fit(X_train, y_train)
```

```python
dum_dum.score(X_test, y_test)
```

```python
d_pred = dum_dum.predict(X_test)
```

```python
sgd_prd = sgd.predict(X_test)
```

```python
lgr_prd = lgr.predict(X_test)
```

```python
confusion_matrix(y_test, sgd_prd)
```

```python
confusion_matrix(y_test, lgr_prd)
```

```python
confusion_matrix(y_test, d_pred)
```

#### Looked at the accuracy score for the Logistic Regression, SGD, and Dummy Classifier models done above. It appears the Logistic Regression model had the more accurate prediction.

```python
print("Accuracy score for Logistic Regression model:\n{:.2f}".format(accuracy_score(y_test, lgr_prd)))
print("Accuracy score for SGD model: \n{:.2f}".format(accuracy_score(y_test, sgd_prd)))
print("Accuracy score for Dummy Classifer model: \n{:.2f}".format(accuracy_score(y_test, d_pred)))
```

#### Looked at the precision scores for the Logistic Regression, SGD, and Dummy Classifier models done above. It appears the Logistic Regression model had the most precise predictions with having the least False Positive values when looking at Age vs. Purchased.

```python
print("Precision score for Logistic Regression model: \n", precision_score(y_test, lgr_prd))
print("Precision score for SGD model: \n", precision_score(y_test, sgd_prd))
print("Precision score for Dummy Classifer model: \n", precision_score(y_test, d_pred))
```

```python
print("Recall score for Logistic Regression model: \n", recall_score(y_test, lgr_prd))
print("Recall score for SGD model: \n", recall_score(y_test, sgd_prd))
print("Recall score for Dummy Classifer model: \n", recall_score(y_test, d_pred))
```

```python
print("Logistic Regression full report\n", classification_report(y_test,lgr_prd))
```

```python
print("Logistic Regression full report\n", classification_report(y_train,lgr.predict(X_train)))
```

```python
print("SGD full report\n", classification_report(y_test,sgd_prd))
```

```python
print("SGD full report\n", classification_report(y_train,sgd.predict(X_train)))
```

```python
print("Dummy full report\n", classification_report(y_test,d_pred))
```

```python
print("Dummy full report\n", classification_report(y_train,dum_dum.predict(X_train)))
```

```python
params = {'n_neighbors': [i for i in range (1, 11)]}
```

```python
grid = GridSearchCV(KNeighborsClassifier(), param_grid=params, scoring = 'accuracy')
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
best = grid.best_estimator_
```

```python
best.fit(X_train, y_train)
```

```python
gridk_prd = best.predict(X_test)
```

```python
confusion_matrix(y_test, gridk_prd)
```

```python
print("Accuracy score for GridSearch KNeighborsClassifier model: \n{:.2f}".format(accuracy_score(y_test, gridk_prd)))
print("Precision score for GridSearch KNeighborsClassifier model: \n", precision_score(y_test, gridk_prd))
print("Recall score for GridSearch KNeighborsClassifier model: \n", recall_score(y_test, gridk_prd))
```

```python
print("KNeighborsClassifier full report\n", classification_report(y_test,gridk_prd))
```

```python
print("KNeighborsClassifier full report\n", classification_report(y_train,best.predict(X_train)))
```
